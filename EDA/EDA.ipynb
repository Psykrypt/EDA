# Importing Libraries and loading the file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Iris=pd.read_csv('Iris.csv')

```

```python
Iris.head()
```

```python
iris=Iris.drop(['Id'],axis=1)
iris
```

# Understanding the data

```python
iris["Species"].value_counts()
```

```python
iris.isnull().sum()
```

# Plotting Graphs to visualize the data

### Countplot

```python
plt.title('Species count')
sns.countplot(iris['Species'])
```

### Scatterplot

```python
plt.figure(figsize=(8,5))
plt.title('Comparison between sepal width and length on the basis of species')
sns.scatterplot(x=iris['SepalLengthCm'],y= iris['SepalWidthCm'], hue = iris['Species'])
```

### Histogram (Univariate Analysis)

```python
sns.FacetGrid(iris,hue="Species").map(sns.distplot,"PetalWidthCm").add_legend()
```

```python
sns.FacetGrid(iris,hue="Species").map(sns.distplot,"SepalLengthCm").add_legend()
```

```python
sns.FacetGrid(iris,hue="Species").map(sns.distplot,"SepalWidthCm").add_legend()
```

```python
sns.FacetGrid(iris,hue="Species").map(sns.distplot,"PetalLengthCm").add_legend()
```

### Pairplot

```python
sns.pairplot(iris,hue="Species")
```

### Boxplot 

```python
sns.boxplot(x="Species",y="PetalLengthCm", data=iris)
```

### Violinplot

```python
sns.violinplot(x="Species",y="PetalLengthCm", data=iris)
```

### Heatmap 

```python
sns.heatmap(iris.corr())
```
